```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = 'training_log/training_log_2.json'
with open(file_path, 'r') as file:
    data = json.load(file)
rewards = np.array(data['Reward'])
distances = np.array(data['Distance'])
steps = np.array(data['Steps'])
```

```python
def moving_average(data, window_size=10):
    return np.convolve(data, np.ones(window_size)/window_size, mode='valid')
```

```python
plt.plot(rewards, alpha=0.5, label='Returns')
rewards_avg = moving_average(rewards)
plt.plot(rewards_avg, label='Moving Average')
plt.xlabel('Episode Number')
plt.ylabel('Episode Returns')
plt.title('Returns Tracking')
plt.legend()
```

```python
plt.plot(distances, alpha=0.5, label='Distance to Goal')
distances_avg = moving_average(distances)
plt.plot(distances_avg, label='Moving Average')
plt.xlabel('Episode Number')
plt.ylabel('Episode Final Distance to Goal')
plt.title('Final Distance to Goal Tracking')
plt.legend()
```

```python
plt.plot(steps, alpha=0.5, label='Number of Steps')
steps_avg = moving_average(steps)
plt.plot(steps_avg, label='Moving Average')
plt.xlabel('Episode Number')
plt.ylabel('Episode Steps Taken')
plt.title('Steps per Episode Tracking')
plt.legend()
```

```python
avg_rewards = rewards / steps
plt.plot(avg_rewards)
```
